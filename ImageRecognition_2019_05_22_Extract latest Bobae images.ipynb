```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import random
import os
```

```python
root = '/Volumes/TriveStorage/code/trive-image-recognition/bobae_latest'
targets = [f'http://www.bobaedream.co.kr/mycar/mycar_view.php?no={x}&gubun=K' for x in range(1_000_000, 2_000_000)]
random.shuffle(targets)

for target in targets:
    result = requests.get(target)
    if result.status_code == 200:
        try:
            soup = BeautifulSoup(result.content)
            title = soup.find_all('h3',class_='tit')[0].get_text()
            image_container = soup.find('div', {'id': 'bx-pager'})
            images = list(set([x['src'].replace('_s1', '') for x in image_container.find_all('img') if 'file4.bobaedream.co.kr/direct' in x['src']]))
            dst = os.path.join(root, title)
            if not os.path.exists(dst):
                os.mkdir(dst)
            for url in images:
                try:
                    response = requests.get(url, stream=True)
                    if not response.ok:
                        continue
                    with open(os.path.join(dst, url.split('/direct/')[1].replace('/', '')), 'wb') as handle:
                        for block in response.iter_content(1024):
                            if not block:
                                break
                            handle.write(block)
                except Exception as e:
                    print(e)
        except Exception as e:
            print(e)
    
    
    
    
```
