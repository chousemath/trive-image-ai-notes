```python
from imageai.Prediction.Custom import CustomImagePrediction
import os
from shutil import copyfile

execution_path = os.getcwd()
# list = os.listdir(os.path.join(execution_path,'cars_refined/test'))
list = os.listdir(os.path.join(execution_path,'frontback'))
prediction = CustomImagePrediction()
prediction.setModelTypeAsResNet()

# print(os.path.join(execution_path,"cars_refined/models/model_ex-001_acc-0.069950.h5"))
prediction.setModelPath(os.path.join(execution_path,"cars_refined/models/model_ex-061_acc-0.959643.h5"))
prediction.setJsonPath(os.path.join(execution_path,"cars_refined/json/model_class.json"))
prediction.loadModel(num_objects=81)


# predictions, probabilities = prediction.predictImage("cars_refined/test/기아_K3_K3(12~15년)/1022509405762742542.jpg", result_count=5)

# print(predictions)
# print(probabilities)
# for eachPrediction, eachProbability in zip(predictions, probabilities):
#   print(eachPrediction + " : " + eachProbability)

total = 0
totalfail = 0

for car in list:
    os.mkdir(execution_path + '/failed_frontback/' + car);
#     carList = os.listdir(execution_path + '/cars_refined/test/' + car)
    carList = os.listdir(execution_path + '/frontback/' + car)
    count = 0
    fail = 0

    print('actual: ' + car)
    for item in carList:
        count += 1
        total += 1
#         imagePath = execution_path + "/cars_refined/test/" + car + "/" + item
        imagePath = execution_path + "/frontback/" + car + "/" + item
        predictions, probabilities = prediction.predictImage(imagePath, result_count=3)
        if (car != predictions[0]):
            print(predictions[0] + " - " + probabilities[0])
#             print('actual: ', car, item, ', predict: ', predictions[0], '-', probabilities[0])
            fail += 1
            totalfail += 1
            copyfile(imagePath, execution_path + '/failed_frontback/' + car + '/' + str(count) + predictions[0] + " - " + probabilities[0])
    print(car,'- total: ', count, 'failed: ', fail)
print('total: ', total, 'total failed: ', totalfail)
```
