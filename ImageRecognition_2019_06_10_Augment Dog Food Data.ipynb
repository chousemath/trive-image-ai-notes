```python
# import the OpenCV package
import cv2
import os
import random
import numpy as np
from PIL import Image
from pprint import pprint
import shutil
import csv
```

```python
def rotate_bound(image, angle):
    # grab the dimensions of the image and then determine the
    # center
    (h, w) = image.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    # grab the rotation matrix (applying the negative of the
    # angle to rotate clockwise), then grab the sine and cosine
    # (i.e., the rotation components of the matrix)
    M = cv2.getRotationMatrix2D((cX, cY), -angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])
    # compute the new bounding dimensions of the image
    nW = int((h * sin) + (w * cos))
    nH = int((h * cos) + (w * sin))
    # adjust the rotation matrix to take into account translation
    M[0, 2] += (nW / 2) - cX
    M[1, 2] += (nH / 2) - cY
    # perform the actual rotation and return the image
    return cv2.warpAffine(image, M, (nW, nH))
```

```python
root = '/Volumes/TriveStorage/code/dog_foods/images'
for dog_food in [x for x in os.listdir(root) if x != '.DS_Store']:
    print(dog_food)
    dog_food_path = os.path.join(root, dog_food)
    print(dog_food_path)
    path_original = os.path.join(dog_food_path, 'original')
    path_90 = os.path.join(dog_food_path, 'rot90')
    path_180 = os.path.join(dog_food_path, 'rot180')
    path_270 = os.path.join(dog_food_path, 'rot270')
    print(path_original)
    print(path_90)
    if not os.path.exists(path_90):
        os.mkdir(path_90)
    print(path_180)
    if not os.path.exists(path_180):
        os.mkdir(path_180)
    print(path_270)
    if not os.path.exists(path_270):
        os.mkdir(path_270)
    for image_name in [x for x in os.listdir(path_original) if x != '.DS_Store']:
        image_path = os.path.join(path_original, image_name)
        try:
            dst = os.path.join(path_90, image_name)
            if not os.path.exists(dst):
                img = cv2.imread(image_path)
                rotated = rotate_bound(img, 90)
                cv2.imwrite(dst, rotated)
        except Exception as e:
            print(e)
            
        try:
            dst = os.path.join(path_180, image_name)
            if not os.path.exists(dst):
                img = cv2.imread(image_path)
                rotated = rotate_bound(img, 180)
                cv2.imwrite(dst, rotated)
        except Exception as e:
            print(e)

        try:
            dst = os.path.join(path_270, image_name)
            if not os.path.exists(dst):
                img = cv2.imread(image_path)
                rotated = rotate_bound(img, 270)
                cv2.imwrite(dst, rotated)
        except Exception as e:
            print(e)
            
```
