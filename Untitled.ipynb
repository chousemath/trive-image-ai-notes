```python
documents = []
y = []
count = 0
targets= []

for key in encar_categories:
    targets.append(key)
    for document in encar_categories[key]:
        documents.append(document)
        y.append(count)
    count += 1
    
```

```python
vectorizer = CountVectorizer(max_features=450, min_df=5, max_df=0.7, stop_words=None)  
X = vectorizer.fit_transform(documents).toarray()
```

```python
tfidfconverter = TfidfTransformer()  
X = tfidfconverter.fit_transform(X).toarray()  
```

```python
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)  
classifier.fit(X, y)  
```
