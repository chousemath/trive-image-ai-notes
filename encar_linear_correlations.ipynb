```python
data = {}

for doc in docs:
  year = doc['year']
  if '(' in year:
    year = year.replace(year[year.find('년')+1:year.find('(')], '')
  else:
    year = year[:year.find('년')+1]
  title = doc['title'].replace(' ', '').replace('/', '').strip()
  name = f'{title}_{year}'
  
  if title not in data:
    data[title] = [doc['price']]
  else:
    data[title].append(doc['price'])
    
  if name not in data:
    data[name] = [doc['price']]
  else:
    data[name].append(doc['price'])

```

```python
final_data = []
for x in data:
  if len(data[x]) > 0:
    final_data.append([x, sum(data[x]) / len(data[x])])

sorted(final_data, key=itemgetter(0))
    
with open('encar_prices.csv', mode='w') as f:
    wr = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for x in final_data:
      wr.writerow(x)
        
files.download('encar_prices.csv') 
```
